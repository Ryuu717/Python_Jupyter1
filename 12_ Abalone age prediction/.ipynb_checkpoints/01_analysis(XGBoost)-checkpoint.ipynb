```python
train = pd.read_csv("train.tsv", sep="\t")
test = pd.read_csv("test.tsv", sep="\t")
sample = pd.read_csv("sample_submit.csv", header=None)
```

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
```

```python
trainX = pd.get_dummies(train.iloc[:, 1:9])
y = train['Rings']

testX = pd.get_dummies(test.iloc[:, 1:])
```

提出結果は119

# XGB

```python
reg = xgb.XGBRegressor()
```

```python
parameters = {'max_depth': [2,4,6], 'n_estimators': [50,100,200]}
```

```python
gcv = GridSearchCV(reg, parameters, cv=5, scoring="neg_mean_squared_error", n_jobs=-1, return_train_score=True, verbose=1)
gcv.fit(trainX, y)
print(gcv.best_params_, gcv.best_score_)
```

```python
pred_train = gcv.predict(trainX)
```

```python
from sklearn.metrics import mean_squared_error as MSE
print("RMSE",MSE(y, pred_train)**0.5)
```

```python
_df = pd.DataFrame({'actual':y, 'pred':pred_train})
_df.plot(figsize=(20,5))
```

```python
pred_test = reg.predict(testX)
```

```python
sample[1] = pred_test
```

```python
sample.to_csv("submit01.csv", header=None, index=None)
```

# RFR

```python
from sklearn.ensemble import RandomForestRegressor as RF

rfr = RF()
```

```python
parameters = {'max_depth': [2,4,6], 'n_estimators': [50,100,200]}
```

```python
gcv = GridSearchCV(rfr, parameters, cv=5, scoring="neg_mean_squared_error", n_jobs=-1, return_train_score=True, verbose=1)
gcv.fit(trainX, y)
print(gcv.best_params_, gcv.best_score_)
```

```python
pred_train = gcv.predict(trainX)
```

```python
from sklearn.metrics import mean_squared_error as MSE
print("RMSE",MSE(y, pred_train)**0.5)
```

```python
pred_test = gcv.predict(testX)
```

```python
sample[1] = pred_test
```

```python
sample.to_csv("submit02.csv", header=None, index=None)
```
