```python
train_2012 = train_merge[train_merge["year"]==2012]
train_2013 = train_merge[train_merge["year"]==2013]
```

```python
train_2012.head()
```

```python
from sklearn.metrics import mean_squared_error as MSE
from sklearn.linear_model import LinearRegression as LR
from sklearn.ensemble import RandomForestRegressor as RF

from sklearn.model_selection import GridSearchCV
import pydotplus
```

```python
rfr = RF()
```

```python
parameters = {"n_estimators": list(range(80,120)), "max_depth": list(range(3,8))}
```

```python
trainX = train_2012[["stage", "home", "away", "stadium","date","week"]]
trainX = pd.get_dummies(trainX)
y = train_2012["y"]
```

```python
gcv = GridSearchCV(rfr, parameters, cv=5, scoring="neg_mean_squared_error", n_jobs=-1, return_train_score=True)
gcv.fit(trainX, y)
```

```python
gcv.best_params_
```

```python
pred = gcv.predict(trainX)

p = pd.DataFrame({"actual":y, "pred":pred})
p.plot(figsize=(15,4))
print("RMSE",MSE(y,pred)**0.5)
```
